```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = np.loadtxt("/content/drive/MyDrive/datasets/plei_reduced_dataset.csv", delimiter=',', skiprows=1, dtype=str)
df = pd.DataFrame(data,columns=['RA', 'DE', 'plx', 'pmRA', 'pmDE', 'Gmag','BP-RP', 'distance']).astype(float)
```

```python
# Create a CMD visualization for pleiades dataset

plt.figure(figsize=([8,10]))
plt.scatter(df['BP-RP'], df['Gmag'], color='dodgerblue',marker='.', s=5, alpha=0.7)
plt.gca().invert_yaxis()
plt.xlabel('BP-RP [mag]', size=12)
plt.ylabel('G [mag]', size=12)
plt.title('Color v. Apparent Magnitude', size=14)
plt.show()
```

```python
# Calculate Absolute Magnitude for CMD Visualization

M = df['Gmag'] - 5 * np.log10(df['distance']) + 5
print('Absolute Magnitude:', M.mean())

plt.figure(figsize=([8,10]))
plt.scatter(df['BP-RP'], M, color='dodgerblue',marker='.', s=5, alpha=0.7)
plt.gca().invert_yaxis()
plt.xlabel('BP-RP [mag]', size=12)
plt.ylabel('Absolute Magnitude [mag]', size=12)
plt.title('Pleiades CMD', size=14)
plt.savefig('plei_cmd.pdf', bbox_inches='tight', pad_inches=0.3)
plt.show()
```

```python
# Isochrone Fitting with Star Cluster Data

#ISOCHRONE 1 DATA EXTRACTION
df_80 = pd.read_csv('/content/drive/MyDrive/datasets/iso80.txt', sep='\s+', comment='#', skip_blank_lines=True, usecols=[28, 29, 30], names=['Gmag', 'BPmag', 'RP_mag'])
df_80['BP-RP'] = df_80['BPmag'] - df_80['RP_mag']


#ISOCHRONE 2 DATA EXTRACTION
df_100 = pd.read_csv('/content/drive/MyDrive/datasets/iso100.txt', sep='\s+', comment='#', skip_blank_lines=True, usecols=[28, 29, 30], names=['Gmag', 'BPmag', 'RP_mag'])
df_100['BP-RP'] = df_100['BPmag'] - df_100['RP_mag']


#ISOCHRONE 3 DATA EXTRACTION
df_125 = pd.read_csv('/content/drive/MyDrive/datasets/iso125.txt', sep='\s+', comment='#', skip_blank_lines=True, usecols=[28, 29, 30], names=['Gmag', 'BPmag', 'RP_mag'])
df_125['BP-RP'] = df_125['BPmag'] - df_125['RP_mag']


#ISOCHRONE 4 DATA EXTRACTION
df_150 = pd.read_csv('/content/drive/MyDrive/datasets/iso150.txt', sep='\s+', comment='#', skip_blank_lines=True, usecols=[28, 29, 30], names=['Gmag', 'BPmag', 'RP_mag'])
df_150['BP-RP'] = df_150['BPmag'] - df_150['RP_mag']
```

```python
# Isochrone Fit with Pleiades Cluster

plt.figure(figsize=(8,10))
plt.plot(df_80['BP-RP'], df_80['Gmag'], color='red', label='Iso1 (80 Myr)', alpha=0.5)
plt.plot(df_100['BP-RP'], df_100['Gmag'], color='green', label='Iso2 (100 Myr)',  alpha=0.5)
plt.plot(df_125['BP-RP'], df_125['Gmag'], color='orange', label='Iso3 (125 Myr)',alpha=0.5)
plt.plot(df_150['BP-RP'], df_150['Gmag'], color='magenta', label='Iso4 (150 Myr)',alpha=0.5)
plt.scatter(df['BP-RP'], M, color='dodgerblue',marker='.', label='Pleiades')
plt.gca().invert_yaxis()
plt.xlabel('BP-RP [mag]', size=12)
plt.ylabel('M [mag]', size=12)
plt.ylim(15, -7)
plt.xlim(-0.5, 5)
plt.title('Pleiades Cluster with Various Isochrone', size=14)
plt.legend()
#plt.savefig('plei_isofit.pdf', bbox_inches='tight', pad_inches=0.3)
plt.show()
```
